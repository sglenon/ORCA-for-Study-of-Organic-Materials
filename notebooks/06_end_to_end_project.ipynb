# 06 — End-to-End Project Planner

Create a calculation manifest before running a ligand or metal-complex project. This notebook turns the workflow into an auditable state tree.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
print("Repository root:", ROOT)
```

```python
import pandas as pd

manifest = pd.DataFrame([
    {"species": "ligand_parent", "charge": 0, "multiplicity": 1, "job": "opt_freq", "status": "planned"},
    {"species": "ligand_radical_site1", "charge": 0, "multiplicity": 2, "job": "opt_freq", "status": "planned"},
    {"species": "ligand_radical_cation", "charge": 1, "multiplicity": 2, "job": "opt_freq", "status": "planned"},
    {"species": "ligand_radical_anion", "charge": -1, "multiplicity": 2, "job": "opt_freq", "status": "planned"},
    {"species": "ligand_deprotonated_site1", "charge": -1, "multiplicity": 1, "job": "opt_freq", "status": "planned"},
    {"species": "ligand_parent", "charge": 0, "multiplicity": 1, "job": "tddft", "status": "blocked_by_opt_freq"},
])
manifest
```

```python
out = ROOT / "notebook_outputs" / "calculation_manifest.csv"
out.parent.mkdir(exist_ok=True)
manifest.to_csv(out, index=False)
print("Wrote", out)
```
