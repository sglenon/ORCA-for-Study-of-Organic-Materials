## State bookkeeping

The parent, radical, radical cation, radical anion, and deprotonated species must each be optimized and frequency-checked with the correct charge and multiplicity.

```python
import pandas as pd
from orca_reviewer.thermochemistry import antioxidant_descriptors, composite_enthalpy

states = pd.read_csv(ROOT / "sample_data" / "antioxidant_enthalpies.csv").set_index("state")
states
```

```python
H = states["enthalpy_Eh"].to_dict()
result = antioxidant_descriptors(
    parent_h=H["parent"],
    radical_h=H["radical_site1"],
    radical_cation_h=H["radical_cation"],
    radical_anion_h=H["radical_anion"],
    deprotonated_h=H["deprotonated_site1"],
    hydrogen_atom_h=H["hydrogen_atom"],
    electron_h=0.0,
    proton_h=0.00236,
    unit="kJ/mol",
)
pd.Series(result)
```

## Composite enthalpy

Use a high-level single-point electronic energy plus the lower-level thermal correction. This avoids counting the lower-level electronic energy twice.

```python
example = composite_enthalpy(
    high_level_sp=-500.250,
    low_level_total_enthalpy=-499.930,
    low_level_electronic_energy=-500.000,
)
print("Composite enthalpy / Eh:", example)
```
