# 01 — Setup and Input Syntax

Build and validate ORCA inputs without needing the ORCA binary. When `ORCA_EXE` is configured, the final cell can run the example.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
print("Repository root:", ROOT)
```

## Build a water input

The builder forces an explicit charge, multiplicity, and coordinate source.

```python
from orca_reviewer.input_builder import OrcaInput

water = OrcaInput(
    method_line="HF def2-SVP TightSCF",
    charge=0,
    multiplicity=1,
    coordinates=[
        "O   0.000000  0.000000  0.000000",
        "H   0.758602  0.000000  0.504284",
        "H  -0.758602  0.000000  0.504284",
    ],
    blocks={"pal": ["nprocs 1"], "scf": ["MaxIter 200"]},
    comments=["First ORCA input", "Closed-shell neutral water"],
)
print(water.render())
```

```python
target = ROOT / "notebook_outputs" / "water.inp"
water.write(target)
print("Wrote", target)
```

## Detect ORCA

The notebook remains useful in dry-run mode. It will not pretend that a calculation ran.

```python
from orca_reviewer.runner import find_orca, shell_command
exe = find_orca()
print("ORCA executable:", exe or "not found — dry-run mode")
print("Command:", shell_command(target))
```
