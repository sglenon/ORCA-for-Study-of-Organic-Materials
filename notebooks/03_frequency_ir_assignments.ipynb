# 03 — Frequencies, IR, and Assignments

Parse frequencies, check for imaginary modes, broaden an IR stick spectrum, and create an assignment worksheet.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
print("Repository root:", ROOT)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from orca_reviewer.parsers import parse_frequencies, parse_ir_spectrum, parse_total_enthalpy
from orca_reviewer.spectra import gaussian_broaden

sample = ROOT / "sample_data" / "sample_freq.out"
freqs = parse_frequencies(sample)
ir = pd.DataFrame(parse_ir_spectrum(sample))
print("Total enthalpy / Eh:", parse_total_enthalpy(sample))
print("Imaginary frequencies:", [f for f in freqs if f < -1.0])
ir
```

```python
scale_factor = 0.97
ir["scaled_frequency_cm-1"] = ir["frequency_cm-1"] * scale_factor
grid = np.linspace(350, 3500, 4000)
y = gaussian_broaden(grid, ir["scaled_frequency_cm-1"], ir["intensity_km_mol"], fwhm=25)

plt.figure(figsize=(10, 4))
plt.plot(grid, y)
plt.gca().invert_xaxis()
plt.xlabel("Wavenumber / cm$^{-1}$")
plt.ylabel("Simulated intensity / a.u.")
plt.title("Synthetic IR spectrum — scaled and Gaussian broadened")
plt.tight_layout()
plt.show()
```

Populate `mode_description` only after animating each mode in a molecular viewer.
